# Implémentation d'un modèle de régression logistique

```python
# On défnit X et y
X = pd.get_dummies(data[['ADR_SECTEUR','DIAMETREARBREAUNMETRE','NOTEDIAGNOSTIC','SOUS_CATEGORIE_DESC','VIGUEUR']])
y = data.loc[:, data.columns == 'DEFAUT']
```

```python
# Suréchantillonage synthétique avec Smote

from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
sm = SMOTE(k_neighbors=3, sampling_strategy=0.75)
X_res, y_res = sm.fit_resample(X_train, y_train)

print("Pourcentage d'arbre avec un défaut sur l'ancien échantillon ",round(pourcentage_avec_defaut,2), "%" ,"\nProportion d'arbre avec un défaut sur le nouvel échantillon", round(len(y_res[y_res['DEFAUT'] == 1])/len(y_res),2)*100, "%" )
```

```python
X_res.head()
```

```python
X_res.shape
```

```python
# Séléctionner les variables les plus pertinantes pour notre régression logistique avec la méthode RFE 

from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

logistique_model = LogisticRegression()
rfe = RFE(logistique_model)
rfe.fit(X_res, y_res)
X_res = X_res[X_res.columns[rfe.support_]]
print(rfe.support_)
print(rfe.ranking_)  
```

```python
X_res.shape 
```

```python
from sklearn.metrics import accuracy_score, f1_score,recall_score,precision_score, confusion_matrix

X_train, X_test, y_train, y_test = train_test_split(X_res , y_res, test_size=0.3)
logreg = LogisticRegression()
logreg.fit(X_train, y_train) 
y_pred = logreg.predict(X_test) 
accuracy_1 = accuracy_score(y_test,y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test,y_pred)
results = pd.DataFrame([['Logistic Regression', accuracy_1,prec,rec,f1]],columns=['Model', 'Accuracy', 'Precision', 'Recall','F1 Score'])
results
```

```python
# Matrice de confusion
cm = confusion_matrix(y_test, y_pred) 
df_cm = pd.DataFrame(cm, index = (0, 1), columns = (0, 1))
plt.figure(figsize = (10,7))
sns.set(font_scale=1.4)
sns.heatmap(df_cm, annot=True, fmt='g')
```

```python
# Validation croisée
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator= logreg, X=X_train, y=y_train, cv=10)
accuracies.mean()
accuracies.std()
print('Logistic Regression Accuracy: %0.3f (+/- %0.3f)' % (accuracies.mean(), accuracies.std() * 2))
```

```python
#Coefficient de la regression
pd.concat([pd.DataFrame(X_train.columns, columns = ["features"]),
           pd.DataFrame(np.transpose(logreg.coef_), columns = ["coef"])
           ],axis = 1)
```

```python
# Comparaison de notre classification avec un classifieur aléatoire

from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
logit_roc_auc = roc_auc_score(y_test, logreg.predict(X_test))
fpr, tpr, thresholds = roc_curve(y_test, logreg.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Regression logistique (air = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taux de faux positifs')
plt.ylabel('Taux de vrai positifs')
plt.legend(loc="lower right")
plt.show()
```

## Comparaison avec un modèle de forêt aléatoire

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
X_train, X_test, y_train, y_test = train_test_split(X_res , y_res, test_size=0.3)
clf=RandomForestClassifier(n_estimators=100)
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
accuracy_2 = metrics.accuracy_score(y_test, y_pred)
```

```python
comparaison_precision = pd.DataFrame([[accuracy_1,accuracy_2]],columns=['Regression Logistique', 'Forêt aléatoire'])
comparaison_precision
```
